#### Importing Modules in python: Modules and packages

In Python, modules and packages help organize and reuse code. Here's a comprehensive guide on  importing modules in Python.


```python
# importing entire module
import math
math.sqrt(15)
```

```python
# when you want to import specific  functions from a module
# This will save memory and improve performance

from math import sqrt,pi
print(math.sqrt(25))
print(sqrt(81))
print(pi)
```

```python
import numpy as np
np.array([1, 2, 3,4,5])
```
