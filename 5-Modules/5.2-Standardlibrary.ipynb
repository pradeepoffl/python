#### Standard Library overview 
Python's Standard Library is a vast collection of modules and packages that come bundled with python,providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in python standard library.

```python
### array module is a .py file with class,function,variables

import array
arr=array.array('i',[1,2,3,4,5])
print(arr)
```

```python
# math module

import math
print(math.sqrt(25))
print(math.pow(2,3))
print(math.pi)
```

```python
## Random module used for pickuping 
import random
print(random.randint(1,10))
print(random.choice(['apple','mango','cherry']))

```

```python
import os

# specify the folder and files name
dir_name = 'Test_directory'
file_name = 'test_file.txt'

# create the directory

os.makedirs(dir_name,exist_ok=True)

# create the file inside the directory / folder
# write/input some data inside a file 

file_path = os.path.join(dir_name,file_name)
with open(file_path,'w') as file:
    file.write('This is a sample file.')

    # successfully File is create inside the directory
    print(f'created folder "{dir_name}" and file "{file_name}" inside it.')

# Read/output/print data inside the file

with open(file_path,'r') as file:
    out_of_file = file.read()
    print(f'Output Filename:{file_name}={out_of_file}')
```

```python
# High level operations on files ans collection of files
import shutil
shutil.copyfile('source.txt','destination.txt') #copying data from one file to otherfile 
```

```python
# Data Serialization

import json
data = {'name':'krish','age':25}
json_str = json.dumps(data)
print(json_str,' == ',type(json_str))

# we will information from API in string formate so we convert it to dictionary
parsed_data = json.loads(json_str) # str -> dict
print(parsed_data,' == ',type(parsed_data))
```

```python
import csv
with open('example.csv',mode='w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name','age'])
    writer.writerow(['krish',32])

with open('example.csv',mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
# datetime module
from datetime import datetime,timedelta

now=datetime.now()
print(now)

yesterday = now - timedelta(days=1)
print(yesterday)
```

```python
# time module

import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
# regular expression
import re
pattern=r'\d+'
text='There are 123 apples'
match=re.search(pattern,text)
print(match.group())

```
