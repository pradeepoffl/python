```python
#21 sum of list elements

def sum_list(numbers):
    # Your code goes here

    sum_of_list = 0
    for iterator in numbers:
        sum_of_list += iterator
    return sum_of_list
    
print(sum_list([1, 2, 3, 4, 5]))
print(sum_list([10, -5, 7, 8, -2]))
```

```python
#23 Remove Duplicate element in list

def remove_duplicates(lst):
    # Your code goes here
    unique_list=[]
    for iterator in lst:
        if iterator not in unique_list:
            unique_list.append(iterator)
    return unique_list
    
    '''
    #2 solution
    unique_list=set(lst)
    return list(unique_list)
    '''
print(remove_duplicates([-1,-1,2,3,4,1,4]))
```

```python
#25 program to reverse a list
'''
def reverse_list(lst):
    # Your code goes here
    result=[]
    for iterator in range(len(lst)-1,-1,-1):
        if lst[iterator] not in result:
            result.append(lst[iterator])
    return result
'''

#2 approach
def reverse_list(lst):
    start=0
    end=len(lst)-1
    while start<end:
        lst[start],lst[end]=lst[end],lst[start]
        start+=1
        end-=1
    return(lst)


print(reverse_list([1, 2, 3, 4, 5]))
```

```python
#26 count even and odd numbers in a list

def count_even_odd(lst):
    # Your code goes here
    even_count=0
    odd_count=0
    for iterator in lst:
        if iterator%2==0:
            even_count+=1 
        else:
            odd_count+=1
    return even_count, odd_count
print(count_even_odd([1,2,3,4,5,5,4]))
```
