```python
#21 sum of list elements

def sum_list(numbers):
    # Your code goes here

    sum_of_list = 0
    for iterator in numbers:
        sum_of_list += iterator
    return sum_of_list
    
print(sum_list([1, 2, 3, 4, 5]))
print(sum_list([10, -5, 7, 8, -2]))
```

```python
#26 count even and odd numbers in a list

def count_even_odd(lst):
    # Your code goes here
    even_count=0
    odd_count=0
    for iterator in lst:
        if iterator%2==0:
            even_count+=1 
        else:
            odd_count+=1
    return even_count, odd_count
print(count_even_odd([1,2,3,4,5,5,4]))
```

```python
#27 two consecutive difference 

def max_consecutive_difference(lst):
    # Your code goes here
    max=0 #edge case if the list is empty or has only one element return 0
    for i in range(1,len(lst)):
        abs_value=abs(lst[i] - lst[i-1])
        if max < abs_value:
            max=abs_value
    return max

print(max_consecutive_difference([12, 20, 13, 4, 5]))
print(max_consecutive_difference([1]))
```

```python
#28 merge two list

def merge_two_sorted_lists(list1, list2):
    # Your code goes here
    f1,f2=0,0
    new_bag=[]
    while f1 < len(list1) and f2 < len(list2):
        if list1[f1] < list2[f2]:
            new_bag.append(list1[f1])
            f1+=1 
        else:
            new_bag.append(list2[f2])
            f2+=1 
    while f1 < len(list1):
        new_bag.append(list1[f1])
        f1+=1 
    while f2 < len(list2):
        new_bag.append(list2[f2])
        f2+=1 
    return new_bag
            


'''
#2 solution
def merge_two_sorted_lists(list1, list2):
    # Your code goes here
    final=list1 + list2
    return sorted(final)
''' 
print(merge_two_sorted_lists(list1 = [1, 3, 5], list2 = [2, 4, 6]))
```

```python
#29 rotate a list

def rotate_list(lst, k):
    # Your code goes here
    if not lst:
        return []
    k=k % len(lst)
    for iterator in range(k):
        last_ele = lst.pop()
        lst.insert(0,last_ele)
    return lst

print(rotate_list([10,20,30,40,50],3))
```

```python
#30 Merging two list into one dictonary

def merge_lists_to_dictionary(keys, values):
    # Your code goes here
    result={}
    
    if len(keys) != len(values): 
        return False
    for i in range(len(keys)):
        result[keys[i]]=values[i]
    
    return result
 
# result=dict(zip(keys,values)) # alternative solution when we have equal number of keys and values if they are short they get fail 

print(merge_lists_to_dictionary(keys = ['a', 'b', 'c'],values = [1, 2, 3]))
```

```python
#31 Merge 3 dictionary into one dictionary

def merge_three_dictionaries(dict1, dict2, dict3):
    # Your code goes here
    #merge={**dict1,**dict2,**dict3}
    merge=dict1.copy()
    for keys,values in dict2.items():
        merge[keys]=values
    for keys,values in dict3.items():
        merge[keys]=values
    return merge

print(merge_three_dictionaries({'a': 1, 'b': 2}, {'c': 3, 'd': 4}, {'e': 5, 'f': 6}))
```

```python
#32 words Frequency in a sentence

def count_word_frequency(sentence):
    # Your code goes here
    result={}
    words=sentence.split()
    for iterator in words:
        if iterator in result:
            result[iterator]+=1
        else:
            result[iterator]=1
    return result

print(count_word_frequency('the quick brown fox jumps over the lazy dog'))
```
