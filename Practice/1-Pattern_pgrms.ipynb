#### Practice pattern programs:
1. Rows: Represent how many lines the pattern has
2. Columns: Represent how many characters in each row

```python
# Triangle

n=6
for i in range(1,n):
    for j in range(i):
        print("*",end="")
    print()
```

```python
# Hill Pattern Triangle

n=5
for i in range(n):

    for j in range(i,n): # 1 Trainagle
        print(" ",end=" ")

    for j in range(i): # 2 Triangle
        print("*",end=" ")
    
    for j in range(i-1):
        print("*",end=" ")

    print()
```

```python
# square parallel bar pattern

n=5
for i in range(n):
    for j in range(n):
        if (j==0 or j==n-1):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# Square plus pattern
n=5
for i in range(n):
    for j in range(n):
        if (i==n//2 or j==n//2):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# Cross Pattern
n=5
for i in range(n):
    for j in range(n):
        if (i==j or i+j==n-1):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# Hollow Square pattern

n=5
for i in range(n):
    for j in range(n):
        if (i==0 or i==n-1 or j==0 or j==n-1):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# Hollow Increasing Triangle

n=5
for i in range(n):
    for j in range(n): 
        if (j==0 or i==n-1 or i==j):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# Hollow Decreasing Triangle

n=5
for i in range(n):
    for j in range(n): 
        if (i==0 or j==0 or i+j==n-1):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# Hollow hill pattern

n=5
for i in range(n):

    for j in range(i,n): # 1 Trainagle
        print(" ",end=" ")

    for j in range(i): # 2 Triangle
        if (i==n-1 or j==0):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    
    for j in range(i+1):
        if (i==n-1 or i==j):
            print("*",end=" ")
        else:
            print(" ",end=" ")

    print()
```
