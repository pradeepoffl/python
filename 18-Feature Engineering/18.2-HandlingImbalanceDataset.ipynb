## Handling Imbalanced Dataset

1. Up Sampling
2. Down Sampling


```python
## Creating sample dataset

import numpy as np 
import pandas as pd 

# set the seed for reproducibility 
np.random.seed(123)

# create a sample dataframe with 2 class

n_samples = 1000 
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
## Create my dataframe with imbalanced dataset

class_0 = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0] * n_class_0

})


class_1 = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[1] * n_class_1

})
```

```python
class_0  ## excle sheet dictionary with key,values
```

```python
class_1
```

1. upsampling

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
df
```

```python
df['target'].value_counts()
```

```python
df_minority=df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
df_majority 
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True,
         n_samples=len(df_majority),
         random_state=42)
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled = pd.concat([df_minority_upsampled,df_majority])
df_upsampled
```

```python
df_upsampled['target'].value_counts()
```

2. Down Sampling

```python
df_minority
```

```python
## Resample imbalanced dataset to downsampling

from sklearn.utils import resample
df_majority_downsample = resample(df_majority,replace=True,
                                  n_samples=len(df_minority),
                                  random_state=42)
df_majority_downsample
```

```python
## concating downsample 
df_downsample = pd.concat([df_minority,df_majority_downsample])
df_downsample['target'].value_counts()
```
