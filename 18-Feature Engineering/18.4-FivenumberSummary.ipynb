#### 5 Number Summary and Box Plot

- Minimum, maxmium, median, Q1, Q3, IQR and outliers are calculated and displayed in a summary table.

```python
import numpy as np
lst_marks =[45,32,56,75,89,90,87,67,54,45,98,99,76,74]
minimum,Q1,median,Q3,maximum= np.quantile(lst_marks,[0,0.25,0.5,0.75,1])
```

```python
minimum,Q1,median,Q3,maximum 
```

```python
IQR = Q3 -Q1
print(IQR)
```

```python
import seaborn as sns
sns.boxplot(lst_marks)
```
