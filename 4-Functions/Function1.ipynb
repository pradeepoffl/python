### Parameter vs Argument

#### Type of arguments

1. Default arguments
2. Positional arguments
3. keyword arguments


```python
# Default arguments
def power(a=5,b=6):
    return a**b
print(power())
print(power(2,4))
print(power(b=2))

# positional arguments
def power(a,b):
    return a**b

#print(power()) # throws error like missing 2 positional argument
print(power(3,4))

## keyword arguments you can pass by name
def power(a,b):
    return a**b

print(power(b=2,a=5))
print(power(a=1,b=5))
```

```python
def func(**kwargs):
    result=[]
    for names,age in kwargs.items():
        result.append((names,age)) #list.append() takes exactly one argument (2 given)
    return dict(result)

print(func(pradeep=24, subbu=24, ajay=22, vishnu=23))
        
```

The return statement in Python is used to exit a function and send back a value (or values) to the caller. It effectively marks the end of the function’s execution and specifies what the function should output. Here’s a breakdown of its characteristics, nature, limitations, and advantages, followed by answers to the specific questions you asked.

Characteristics and Nature of return
Terminates Function Execution: When return is executed, the function stops running. Any code after the return statement within the function will not be executed.

Returns a Value: The return statement can return any type of data (integers, strings, lists, dictionaries, etc.) or even multiple values (usually as a tuple if you separate them by commas).

Optional: If a function does not have a return statement, or if return is used without a value, the function will return None by default.

Supports Multiple Values: Python allows returning multiple values from a function, which are returned as a tuple.

Advantages of return
Flexibility: You can return any type of data from a function, making it very versatile.
Reusability: By returning values, functions can be reused in different contexts, making your code more modular and easier to maintain.
Control Over Output: The return statement gives you precise control over what is sent back to the caller, which is crucial for writing functions that perform specific tasks and return specific results.
Performance: Returning a value and terminating the function can improve performance since no further unnecessary code is executed.
Limitations of return
Single Exit Point: Once return is executed, the function exits, meaning you can’t execute any further code within the function.
One return Statement per Path: While you can use multiple return statements within a function, only one of them will be executed, depending on the control flow (e.g., within different branches of an if statement).
Complexity with Multiple Returns: Functions with multiple return statements can sometimes be harder to read and maintain, especially if they are scattered throughout the function.
1. Why Use return?
To Provide Output: The primary reason to use return is to provide an output from a function. This allows the function to pass data back to the caller, making it useful for a wide range of tasks, from simple calculations to complex data processing.

To End Function Execution: Sometimes, you want to terminate a function early under certain conditions. The return statement can be used to exit the function as soon as a result is determined, improving the efficiency of the code.

2. When to Use return?
When You Need a Result: Use return whenever you want your function to provide a result that will be used elsewhere in your code. For example, in mathematical computations, data processing, or functions that fetch data from a database or an API.

When Termination Is Required: Use return to end the execution of a function based on certain conditions, which is particularly useful in functions with conditional logic.

3. Where to Use return?
In Any Function: You can use return in any function where you want to send a value back to the caller. It’s especially common in functions that are designed to compute a value, retrieve data, or make decisions based on input.

Inside Loops or Conditionals: Often, return is used inside loops or conditional statements within a function to return a value as soon as a particular condition is met.

```python
L=[1,2,3,4,5]

#General what we do is we will write a if and verified value is greater than
# we append the values to list

result=list(filter(lambda x:x>3,L))
print(result)


fruits=["apple","pineapple","pears","orange"]
result=list(filter(lambda x:x.startswith('p'),fruits))
print(result)
```
