```python
#multiple inheritance:

class A:
    def __init__(self,name):
        self.name=name

    def get_name(self):
        return f'My name is {self.name}'

class B:
    def __init__(self,age):
        self.age=age

    def get_age(self):
        return f'age is {self.age}'

class C(A,B):
    def __init__(self,name,age,phone_number):
        self.no=phone_number
        A.__init__(self,name)
        B.__init__(self,age)

    def student_details(self):
        print(f'My name is {self.name}, age is {self.age} and phone number is {self.no}')

#creating object for class A and calling it
obj_a = A("pradeep")
print(obj_a.get_name())

# creating object for class B and calling it

obj_b = B(24)
print(obj_b.get_age())

# creating object for class c and calling it

obj_c = C('pradeep',24,999585)
obj_c.student_details()
```
