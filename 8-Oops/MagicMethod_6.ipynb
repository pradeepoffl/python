#### Magic Methtods
Magic method in python, also know as dunder methods (double underscore methods), are special methods that start and end with double underscores. These methods enable you to define the behavior of objects for built-in operators, such as arithmetic operations, comparsions and more.

Magic methods are predefined methods in python that you can override to change the behaviour of your objects. some common magic methods includes:

Special Methods in Python:

1. **`__init__`**: Initializes a new instance of a class.
2. **`__str__`**: Returns a string representation of an object.
3. **`__repr__`**: Returns an official string representation of an object.
4. **`__len__`**: Returns the length of an object
