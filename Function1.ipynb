```python
def odd_even(num):
    '''  
    This function named odd_even & returns odd & even with respect to number given by user
    Input: num
    output: odd & even 
    created on 11th may 2024
    Handling exception condition in real time like 
    when a user enter other data type
   
    Q/A: what is difference the parameter and argument?
    when you giving input in function it called parameter ex: def odd_even(num):
    during calling of function its called argument ex odd_even(10);
    Parameter is place holder for argument 
    '''
    if type(num)==int:
        if num%2==0:
            print(f'Even number:={num}')
        else:
            print(f'Odd number:={num}')
    else:
        return "Try with number"

odd_even(10)
odd_even(13)
print(odd_even.__doc__)
odd_even('pradeep')
```

### Parameter vs Argument

#### Type of arguments

1. Default arguments
2. Positional arguments
3. keyword arguments


```python
# Default arguments
def power(a=5,b=6):
    return a**b
print(power())
print(power(2,4))
print(power(b=2))

# positional arguments
def power(a,b):
    return a**b

#print(power()) # throws error like missing 2 positional argument
print(power(3,4))

## keyword arguments you can pass by name
def power(a,b):
    return a**b

print(power(b=2,a=5))
print(power(a=1,b=5))
```

### args & Kwargs

*args and **kwargs are special python keywords that are used
 to pass variable lenght of arguments to a function 

*args-allows us to pass a variable number of non-keywords arguments to a function

**kwargs-allows us to pass any number of keyword arguments



```python
"""  
when user want to input 1,2,3.. arguments its fine what about when user want 
to input 100-200 arguments that where *args are use, type of *args is tuple.

"""

def adds(a,b,c):
    return a+b+c
adds(1,2,3)
```

```python
def adds(*args):  # *args instead u can use *input or anything
    output=0
    for i in args:
        output+=i
    print(type(args))
    print(args)
    return output
adds(1,2,4,5,6,8,8,9)
        
```

```python
def display_info(**kwargs):
    print(type(display_info))
    print(kwargs)

    for k,v in kwargs.items():
        print(k,v)
    
#display_info('apple'120,'orange':240,'greenapple':150) -syntax error
display_info('apple'=120,'orange'=240,'greenapple':=50) # mistakes debug

```

```python
def display_info(**kwargs):
    print(type(kwargs))
    print(kwargs)

    for k,v in kwargs.items():
        print(k,'--',v)
    
display_info(apple=120,orange=150,greenapple=220)

```
