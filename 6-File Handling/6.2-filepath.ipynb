Working on File Handling
 

```python
# create a new directory
import os
new_directory = "package"
os.mkdir(new_directory)
print(f"Directory'{new_directory}'create")
```

```python
## listing files and directories
items=os.listdir('.') # dot represent cwd 
print(items)
```

```python
## joining paths

dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(dir_name,file_name)
print(full_path)  # Output: folder/file.txt
```

```python

dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(os.getcwd(),dir_name,file_name)
print(full_path)  # Output: folder/file.txt
```

```python
path = 'example.txt'
if os.path.exists(path):
    print(f'path exist: {path} ')
else:
    print("Doesn't exists")
    
```

```python
# checking if a path is a file or directory

import os
path = 'example.txt' #  replace with your path 'package'

if os.path.isfile(path):
    print(f'path exist: {path} is a file')
elif os.path.isdir(path):
    print("its directory")
else:
    print("Doesn't exist")
    
```

```python
## Getting the absolute path

relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print(absolute_path)  # Output: /path/to/example.txt
```
