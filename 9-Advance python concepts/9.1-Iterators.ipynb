#### Iterators:
Iterators are advanced python concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
mylist=[1,2,3,4,5]
for i in mylist:
    print(i)
```

```python
print(type(mylist))
print(mylist)
```

```python
## Iterator 
iterator=iter(mylist) # using iter() 
print(type(iterator))
print(iterator) # it will not give exposure to complete structure
# here it use lazyloading techique
```

```python
## iterate through all the elements
next(iterator)
```

```python
iterator = iter(mylist)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("All values in list have been printed and the list is empty.")
```
