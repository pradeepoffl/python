#### 2. Generator:
What is a Generator?

Imagine you have a big box of toys, but instead of taking all the toys out at once, you take them out one by one when you want to play. A generator in Python works in a similar way. It lets you create a sequence of values, but instead of creating all of them at once and using a lot of memory, it generates each value one at a time when you need it.

Where is it Used?

Generators are used when:
You have a lot of data to handle, like reading a big file or streaming data from the internet.
You want to save memory because creating all items at once can be wasteful.
You want to create an infinite series of values, like counting forever!

When to Use Generators?

You might use generators when:
You need to process items one at a time (like going through a list of names).
You want to make your code faster and more efficient.

Real World Scenario:

Think about a situation where you’re reading a long story from a book. Instead of trying to remember the whole story at once, you read one page at a time. Generators do this with data: they let you "read" (or use) one piece at a time, which is easier and doesn’t use too much memory!

```python
# Example for Generator:

def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # This will return the current count and pause the function
        count += 1   # This will run when the next value is requested

# Using the generator
counter = count_up_to(5)

print(counter)

for number in counter:
   print(number)

#Defining the Generator: The count_up_to function is our generator. It will yield (give) numbers one at a time.
#Using yield: When we use yield, it returns a value and pauses the function, so the next time we ask for a value, it picks up right where it left off.
#Looping through the Generator: We create a generator called counter that will count up to 5, and then we use a for loop to print each number one at a time.
```

#### give me difference between iterator and generator in python in realworld scenario?


Definitions
Iterator: An iterator is an object that implements the iterator protocol, which consists of two methods: __iter__() (which returns the iterator object itself) and __next__() (which returns the next value from the sequence). Iterators can traverse through all the elements of a collection (like a list or a set).

Generator: A generator is a special type of iterator that is defined using a function and uses the yield keyword to produce values one at a time. Generators are easier to create and can produce items on-the-fly, saving memory.

Differences with Real-World Scenarios
Creation:

Iterator: Imagine you have a recipe book. You can create a bookmark (iterator) to keep track of your place. You have to manually create this bookmark and follow it page by page.
Generator: Now, think of a magic cookbook that gives you one recipe at a time whenever you ask for it (using yield). You don’t need to keep a bookmark; you just request the next recipe when you're ready.
Memory Usage:

Iterator: If you create an iterator for a huge list of ingredients, the entire list is stored in memory. It’s like having a giant pantry where all the ingredients are stored at once.
Generator: A generator would only provide one ingredient at a time. It’s like having a chef who only pulls out the ingredient you need when you ask for it, so the pantry isn’t overwhelmed.
State:

Iterator: An iterator remembers its state explicitly. If you want to go back to a previous ingredient, you have to remember or reset the iterator.
Generator: A generator automatically keeps track of where it left off. When you ask for the next ingredient, it picks up right where it stopped, like a chef who remembers the last ingredient you requested.
Performance:

Iterator: Using an iterator for a huge dataset may slow down your program because it loads everything into memory.
Generator: A generator can be faster for large datasets since it generates values on-the-fly, allowing for better performance with less memory.
Example Code



```python
# Iterator example:

class MyIterator:
    def __init__(self, max):
        self.max = max
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.max:
            self.current += 1
            return self.current
        else:
            raise StopIteration

# Using the iterator
iterator = MyIterator(5)
for num in iterator:
    print(num)
```

```python
# Generator:
def my_generator(max):
    count = 0
    while count < max:
        count += 1
        yield count

# Using the generator
for num in my_generator(5):
    print(num)
```

Summary:

Iterators are more general-purpose and require more setup but are useful when you need to maintain more complex state or behavior.

Generators are simpler to create and are ideal for producing a sequence of values one at a time, making them a great choice for many applications where memory efficiency and ease of use are important.



What is a Decorator?
Think of a decorator as a special kind of wrapping paper that you put around a gift to make it look nicer. In Python, a decorator is a way to add extra features or behavior to a function without changing its original code.

Why Use Decorators?
Decorators help us:

Add functionality: You can add new features to existing functions.
Keep code clean: Instead of rewriting the same code, you can wrap it in a decorator.
Make code reusable: You can use the same decorator on multiple functions.
Where and When are Decorators Used?
You might use decorators in real-world scenarios like:

Logging: You want to keep track of when functions are called, how long they take, or what inputs they get.
Access control: You might want to check if a user has permission to run a function.
Caching results: If a function is slow, you can save (or cache) the result so it doesn’t have to compute it again for the same input.

#### 3.Decorator

What is a Decorator?

Think of a decorator as a special kind of wrapping paper that you put around a gift to make it look nicer.
In Python, a decorator is a way to add extra features or behavior to a function without changing its original code.

Why Use Decorators?

Decorators help us:
Add functionality: You can add new features to existing functions.
Keep code clean: Instead of rewriting the same code, you can wrap it in a decorator.
Make code reusable: You can use the same decorator on multiple functions.

Where and When are Decorators Used?

You might use decorators in real-world scenarios like:
Logging: You want to keep track of when functions are called, how long they take, or what inputs they get.

Access control: You might want to check if a user has permission to run a function.

Caching results: If a function is slow, you can save (or cache) the result so it doesn’t have to compute it again for the same input.

```python
#Decorator:

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function runs!")
        func()  # Call the original function
    return wrapper

@my_decorator  # Using the decorator
def say_hello():
    print("Hello!")

# Calling the decorated function
say_hello()
```

code explanation:

Creating the Decorator: We define a decorator called my_decorator that takes a function func as an argument.
Defining the Wrapper: Inside the decorator, we define a wrapper function that adds some extra behavior (printing a message).
Calling the Original Function: The wrapper function calls the original function (func()) after doing something else.
Using the Decorator: We use the @my_decorator syntax above the say_hello function to apply the decorator.
Running the Function: When we call say_hello(), it first prints the message and then runs the original function.
Real World Scenario
Imagine you have a bakery, and you want to make sure every time you bake a cake, you check if you have all the ingredients. Instead of writing the check every time, you could create a decorator that does the checking for you. Then, whenever you bake a cake, you just use that decorator, and it automatically checks the ingredients first!

Summary

A decorator is like wrapping paper for functions. It lets you add new features without changing the original function.
You use decorators to log information, control access, or cache results.
They help keep your code clean and reusable!
