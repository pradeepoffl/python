Logging with Multiple Loggers

You can create multiple loggers for different parts of your application.


```python
import logging

## Create a logging for module 1

logger1=logging.getLogger("Module 1")
logger1.setLevel(logging.DEBUG)

## create a logging for module 2

logger2=logging.getLogger("Module 2")
logger2.setLevel(logging.WARNING)


# Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s'
)
```

```python
## Log messages with different loggers

logger1.debug("This a debug message")
logger2.warning("This is an warning message")
logger1.error("This error message")
```
