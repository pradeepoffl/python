#### Inheritance :

Inheritance is as fundamental concept in object-oriented programming that allow a class to inherit attributes and methods from another class.
This lesson covers single inheritance and multiple inheritance demonstrating how to create and use them in python

```python
# inheritance
# parent class -> child class 


class car(): #Parent class
    def __init__(self,windows,doors,enginetype): # positional arguments 
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype 
    
    # instance method
    def drive(self):
        print(f'I am driving {self.enginetype} car')

```

```python
car1=car(4,5,'petrol')
print(car1.drive())
```

```python
# child class implementation
# single inheritance

class Bmw(car):
    def __init__(self,widows,doors,enginetype,Is_racecar):
        super().__init__(widows,doors,enginetype)
        self.Is_racecar =Is_racecar



    def selfdriving(self):
        print(f'This BMW is {self.Is_racecar} car')

car2=Bmw(4,5,'hyrogen','Racing')
print(car2.selfdriving()) # calling method car2.selfdriving() 
                          # car2.selfdriving resulting this <bound method Bmw.selfdriving of <__main__.Bmw object at 0x0000022D9AF12960>>
```

```python
# Mutliple inheritance
# when a class inherits from more than one base class

# Parent class 1 or base class 1

class Animal:
    def __init__(self,name):
        self.name=name
    
    def animalname(self):
        print("the animal name is {}".format(self.name))
        print(f'Your animal name is {self.name}')
    
# parent class 2 or base class 2

class pet:
    def __init__(self,owner):
        self.owner = owner

#Derived class
class Dog(Animal,pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        pet.__init__(self,owner)

    def speak(self):
        return f'{self.name} say woof'
    

# create object 
animal1=Animal("dog")
print(animal1.name)
print(animal1.animalname())

animal2=pet("raju")
print(animal2.owner)

animal3=Dog("jimmy","abhinaye")
print(animal3.speak())

```
