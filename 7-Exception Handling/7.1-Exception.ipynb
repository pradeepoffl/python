Understanding Excetions:

Exception in python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program. This section will cover the basics of exceptions, including how to use try,except,else and finally blocks.

What are Exceptions?

Exception are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution. common exception includes:
-  ZeroDivisionError: Dividing by zero.
- FileNotFoundError: File not found
- ValueError: Invalid value
- TypeError: Invalid data type


```python
a=b
```

```python
# error is raised in try block and in except block we will handle  the error so that execution is not stopped
try:
    a=b
except:
    print("Error: a and b are not equal")
```

```python
# error is raised in try block and in except block we will handle  the error so that execution is not stopped
try:
    a=b
except NameError as e:
    print(f"NameError: {e}")
```
