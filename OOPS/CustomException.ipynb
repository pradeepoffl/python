#### Custom exception (Raise and Throw an Exception)

```python
class Error(Exception):
    pass

class dobExecption(Error):
    pass

year = int(input("Enter your BOD year:"))
age = 2024 - year

try:
    if age <30 and age>20:
        print("Conguralution, eligible for job")
    else:
        raise dobExecption
except dobExecption:
    print("sorry, your not eligible")
```
