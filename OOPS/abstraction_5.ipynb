# Abstraction 
Abstraction is the concepts of hiding the complex implementation details and showing only the necessary features of an object. 
This helps in reducing programming complexity and efforts.

Examples: washing machine, tv remote, Laptop ..etc!!


```python
from abc import ABC,abstractmethod

class vehicle(ABC): # abstract base class
    def drive(self):
        print("The vehicle is used for driving")
    
    @abstractmethod # we are writing an a abstractmethod so that we can implement this method in your own ways
    def start_engine(self):
        pass

class Car(vehicle): #child class
    def start_engine(self):
        print("car engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()


car=Car()
operate_vehicle(car) # willnot get none 
print(operate_vehicle(car)) # will get none when we use def in print and for calling also using in print 
# that time you will get none 

```
