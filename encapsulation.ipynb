#### Encapsulation and Abstraction 

Encapsulation and abstraction are two fundamental principles of object oriented programming (OOP) that help in designing robust, maintainable and reusable code.
Encapsulation invloves bundling data and methods that operate on the data within a single unit, while abstraction involves hiding complex implementation details
and exposing only the necessary features.

#### Encapsulation 
Encapsultation is the concept of wrapping data(variables) and methods (functions) together as a single unit.
It resricts direct access to some of the object's components, which is a means of preventing accidental interference and misuse of the data 

```python
# Encapsulation with Getter and Setter Methods
# Public, Private, Protected variables or access modifiers 

class Person:
    def __init__(self,name,age): 
        self.name=name    # Public variables
        self.age=age      # Public variables

person=Person("pradeep",24)
print(person.name,person.age) #  directly accessing parameter of a class using object

def get_name(person):
    return person.name

person=Person("ratna tata",86)
print(get_name(person))

dir(person) #Public variables
```

```python
# Private variables ( Double underscore __name)
# Private variable cannot be accessed outside the class
# For accessing inside the class we will be using getter and setter method

class Person:
    def __init__(self,name,age,gender): 
        self.__name=name    # Private variables
        self.__age=age      # Private variables
        self.gender=gender

def get_name(person):
    return person.__name # return person._Person__name (run)

person=Person("ironman",36,"MAle")
print(get_name(person)) 


dir(person) # private variables '_Person__age'
```

```python
# Protected variables are created with single underscore (exmaple:_name)

class Person:
    def __init__(self,name,age,gender): 
        self._name=name    # Protected variables
        self._age=age      # Protected variables
        self.gender=gender
class Employee(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)


employee = Employee("nishith",24,"Male")
print(employee._name)
```

```python
# Encapsulation with Getter and Setter

class Person:
    def __init__(self,name,age):
        self.__name=name # private access modifier or variable
        self.__age=age #private variable

# getter method for name to access inside the class 
    def get_name(self):
        return self.__name
    def get_age(self):
        return self.__age


# setter method for name change the value

    def set_name(self,name):
        self.__name=name

    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("Age cannot be negative") 

person=Person("Krish",34)

# access and modify private variables using getter and setter

print(person.get_name())
print(person.get_age())


person.set_age(35)
print(person.get_age())
person.set_age(-5)
```
